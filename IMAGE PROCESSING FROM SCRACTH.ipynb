## Q3: For the image shown in figure 3, perform different types of thresholding operations available in OpenCV. Understand their purpose and working. For each technique show their output using Matplotlib. Give a title to each image the name technique.

```python
img = cv.imread('coins.jpg')
```

### Orignal Image

```python
# for maintain orignal quality
plt.imshow(img,'gray',vmin=0,vmax=255)
plt.title("Orignal")
```

### BINARY THRESHOLD

```python
# img act as input
# 127 is thrashold
# 255 is maximum value which assign
# if pixel value > 127 than assign 255 else assign 0
_,BINARY_IMG = cv.threshold(img,127,255,cv.THRESH_BINARY)
plt.imshow(BINARY_IMG,'gray',vmin=0,vmax=255)
plt.title("BINARY THRESHOLD")
```

### BINARY INVERS THRESHOLD

```python
# img act as input
# 127 is thrashold
# 255 is maximum value which assign
# if pixel value > 127 than assign 0 else assign 255
_,BINARY_INV_IMG = cv.threshold(img,127,255,cv.THRESH_BINARY_INV)
plt.imshow(BINARY_INV_IMG,'gray',vmin=0,vmax=255)
plt.title("BINARY INVERS THRESHOLD")
```

### Truncate Threshold

```python
# img act as input
# 127 is thrashold
# 255 is maximum value which assign
# if pixel value > 127 than assign 127 else assign pixel value
_,TRUNC_IMAGE = cv.threshold(img,127,255,cv.THRESH_TRUNC)
plt.imshow(TRUNC_IMAGE,'gray',vmin=0,vmax=255)
plt.title("Truncate Threshold Threshold")
```

### TOZERO THRASHOLD

```python
# img act as input
# 127 is thrashold
# 255 is maximum value which assign
# if pixel value > 127 than assign Pixel value else assign 0
_,TOZERO_IMAGE = cv.threshold(img,127,255,cv.THRESH_TOZERO)
plt.imshow(TOZERO_IMAGE,'gray',vmin=0,vmax=255)
plt.title("Truncate Threshold Threshold")
```

### TOZERO_INVERS THRASHOLD

```python
# img act as input
# 127 is thrashold
# 255 is maximum value which assign
# if pixel value > 127 than assign 0 value else assign Pixel value
_,TOZERO_INV_IMAGE = cv.threshold(img,127,255,cv.THRESH_TOZERO_INV)
plt.imshow(TOZERO_INV_IMAGE,'gray',vmin=0,vmax=255)
plt.title("Truncate Threshold Threshold")
```

## Q4: Thermal cameras create an image using infrared radiation. They are useful in obtaining figure even in absence of white light (visible). Figure 4 a sample image obtained from a thermal camera. Apply some digital image processing techniques to segment those objects (people) from the background.

```python
path = 'thermal.jpg'
# print in gray scale
img = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
plt.imshow(img,cmap='gray')
plt.axis('off')
plt.show()
```

### Bitwise_not Method

```python
# filp all pixel values means pixel>=255 -> 0 and pixel==0->255
img1 = cv2.bitwise_not(img)
plt.imshow(img1,cmap='gray')
plt.axis('off')
plt.show()
```

### Inverse Method

```python
# find invers
img1 = 255 - img
# define loop for fram
fig, axs = plt.subplots(1,2,figsize=(12,12))
# presented orignal image
axs[0].imshow(img,cmap='gray')
axs[0].axis('off')
axs[0].set_title('Original Image',fontsize='medium')
# presented inverse image
axs[1].imshow(img1,cmap='gray')
axs[1].axis('off')
axs[1].set_title('Negative Image',fontsize='medium')
plt.show()
```

## Q5: Write code to segment the outer area of the brain (white matter) as shown in figure 5 and display the segmented image side by side with the original image using matplotlib. Give the segmented area a red color and the display image must be a color image

```python
# read image
image = cv2.imread('ctscan.tif')
```

```python
# convort RGB to GRAY
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
# for conversion BGR TO RGB
plt.imshow(cv2.cvtColor(gray, cv2.COLOR_BGR2RGB))
#plt.imshow(gray)
plt.axis("off")
plt.show()
```

```python
# contours for find boundary points
# hierarchy for find solod boundary of object exist in image
# cv2.RETR_TREE connect hirarchy points
# CHAIN_APPROX_SIMPLE for returns only the endpoints that are necessary for drawing the contour line. 
contours, hierarchy =  cv2.findContours(gray,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
# Assign red color on selected boundary 
image = cv2.drawContours(image, contours, -1, (0, 0, 255), 3)
# for matplotlib convert BGR to RGB
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
#plt.imshow(hierarchy)
plt.axis("off")
plt.show()
```

```python
# deifn fram for presentation of image
fig, axs = plt.subplots(1,2,figsize=(12,12))
# for orignal image
axs[0].imshow(cv2.cvtColor(gray, cv2.COLOR_BGR2RGB))
axs[0].axis('off')
axs[0].set_title('Original Image',fontsize='medium')
# for colored image
axs[1].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
axs[1].axis('off')
axs[1].set_title('Colored Image',fontsize='medium')
plt.show()
```
